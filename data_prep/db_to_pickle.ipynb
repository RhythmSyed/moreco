```python
import numpy as np
import pandas as pd
import sqlite3
import os
```

```python
def get_conn():
    db_name = './visual/db/movie_sqlite.db'
    conn = sqlite3.connect(db_name)
    return conn
```

```python
conn = get_conn()
```

```python
tables = [('scores', 'fk_id'),
          ('movies', 'id'),
          ('posters', 'id'), 
          ('tags', 'id'),
          ('trailers', 'id'),
          ('movie_meta', 'id')]
```

```python
# write data out to compressed pickles
for table, index_col in tables:
#     table = 'tags'
    sql = f"""
    select * from {table};
    """
    df = pd.read_sql(sql.format(table), conn, index_col=index_col)
    print(table)
    print()
    print(df.head())
    os.makedirs('./visual/db/pickles/', exist_ok=True)
    df.to_pickle(f'./visual/db/pickles/{table}.pkl', compression='gzip')
```

```python
# test that files can be read in properly
for table, index_col in tables:
    df2 = pd.read_pickle(f'./visual/db/pickles/{table}.pkl', compression='gzip')
    print(table)
    print()
    print(df2.head())
```
